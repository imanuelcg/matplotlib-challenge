```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
df_mouse = pd.read_csv(mouse_drug_data_to_load)
df_clinical = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
lab_data = pd.merge(df_mouse,df_clinical, on='Mouse ID')

# Display the data table for preview
lab_data.head()

```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor = lab_data.groupby(by=['Drug','Timepoint']).mean()['Tumor Volume (mm3)']
# Convert to DataFrame
tumor_mean_df = tumor.to_frame().reset_index()
# Preview DataFrame
tumor_mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_error = lab_data.groupby(by=['Drug','Timepoint']).sem()['Tumor Volume (mm3)']

# Convert to DataFrame
df_tumor_error = tumor_error.to_frame().reset_index()

# Preview DataFrame
df_tumor_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_tumor_mean = tumor_mean_df.pivot(index = 'Timepoint',columns = 'Drug', values = 'Tumor Volume (mm3)')

reformat_tumor_sem = df_tumor_error.pivot(index = 'Timepoint',columns = 'Drug', values = 'Tumor Volume (mm3)')
# Preview that Reformatting worked
reformat_tumor_mean.head()
```

```python
# Generate the Plot (with Error Bars)

#Graphing the line and the error lines
plt.errorbar(reformat_tumor_mean.index,reformat_tumor_mean["Capomulin"], reformat_tumor_sem["Capomulin"], color = "red", 
                       marker = "o",linestyle="dotted",linewidth=0.5, label = "Capomulin")
plt.errorbar(reformat_tumor_mean.index,reformat_tumor_mean["Infubinol"], reformat_tumor_sem["Infubinol"], color = "blue", 
                       marker = "^",linestyle="dotted",linewidth=0.5, label = "Infubinol")
plt.errorbar(reformat_tumor_mean.index,reformat_tumor_mean["Ketapril"], reformat_tumor_sem["Ketapril"], color = "green", 
                       marker = "s",linestyle="dotted",linewidth=0.5, label = "Ketapril")
plt.errorbar(reformat_tumor_mean.index,reformat_tumor_mean["Placebo"], reformat_tumor_sem["Placebo"], color = "black", 
                       marker = "d",linestyle="dotted",linewidth=0.5, label = "Placebo")

plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.hlines([np.arange(35,75,5)],0,45, alpha = 0.25)
plt.legend(loc="upper left")
                                                                    
# Save the Figure
plt.savefig("Tumor Response to Treatment")
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_mean = lab_data.groupby(by=['Drug','Timepoint'])['Metastatic Sites'].mean()

# Convert to DataFrame
metastatic_mean_df = metastatic_mean.to_frame().reset_index()
# Preview DataFrame
metastatic_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_error = lab_data.groupby(by=['Drug','Timepoint'])['Metastatic Sites'].sem()

# Convert to DataFrame
metastatic_error_df = metastatic_error.to_frame().reset_index()
# Preview DataFrame
metastatic_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_mean_pivot = metastatic_mean_df.pivot(index = 'Timepoint',columns = 'Drug', values = 'Metastatic Sites')

metastatic_error_pivot = metastatic_error_df.pivot(index = 'Timepoint',columns = 'Drug', values = 'Metastatic Sites')
# Preview that Reformatting worked
metastatic_mean_pivot.head()
```

```python
# Generate the Plot (with Error Bars)

#Graphing the line and the error lines
plt.errorbar(metastatic_mean_pivot.index,metastatic_mean_pivot["Capomulin"], metastatic_error_pivot["Capomulin"], color = "red", 
                       marker = "o",linestyle="dotted",linewidth=0.5, label = "Capomulin")
plt.errorbar(metastatic_mean_pivot.index,metastatic_mean_pivot["Infubinol"], metastatic_error_pivot["Infubinol"], color = "blue", 
                       marker = "^",linestyle="dotted",linewidth=0.5, label = "Infubinol")
plt.errorbar(metastatic_mean_pivot.index,metastatic_mean_pivot["Ketapril"], metastatic_error_pivot["Ketapril"], color = "green", 
                       marker = "s",linestyle="dotted",linewidth=0.5, label = "Ketapril")
plt.errorbar(metastatic_mean_pivot.index,metastatic_mean_pivot["Placebo"], metastatic_error_pivot["Placebo"], color = "black", 
                       marker = "d",linestyle="dotted",linewidth=0.5, label = "Placebo")

plt.title("Metastatic Spread During Treatment")
plt.ylabel("Treatment Duration (Days)")
plt.xlabel("Met. Sites")
plt.hlines([np.arange(0,3.5,0.5)],0,45, alpha = 0.25)
plt.legend(loc="upper left")
                                                                    
# Save the Figure
plt.savefig("Metastatic Spread During Treatment")
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse = lab_data.groupby(by=['Drug','Timepoint'])['Mouse ID'].count()

# Convert to DataFrame
id_to_count = mouse.to_frame()
mouse_count = id_to_count.rename(columns={"Mouse ID":"Mouse Count"}).reset_index()

# Preview DataFrame
mouse_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_rate_pct = mouse_count.pivot_table('Mouse Count',['Timepoint'],'Drug')
# Preview that Reformatting worked
survival_rate_pct.head()
```

```python
#Create the graph for the survival rate
plt.errorbar(survival_rate_pct.index, (survival_rate_pct['Capomulin']/25) * 100, color = "red", 
                       marker = "o",linestyle="dotted",linewidth=0.5, label = "Capomulin")

plt.errorbar(survival_rate_pct.index, (survival_rate_pct['Infubinol']/25) * 100, color = "blue", 
                       marker = "^",linestyle="dotted",linewidth=0.5, label = "Infubinol")

plt.errorbar(survival_rate_pct.index, (survival_rate_pct['Ketapril']/25) * 100, color = "green", 
                       marker = "s",linestyle="dotted",linewidth=0.5, label = "Ketapril")

plt.errorbar(survival_rate_pct.index, (survival_rate_pct['Placebo']/25) * 100, color = "black", 
                       marker = "d",linestyle="dotted",linewidth=0.5, label = "Placebo")

plt.title("Survival During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time (Days)")
plt.grid()
plt.legend(loc="lower left")
                                                                    
# Save the Figure
plt.savefig("Survival During Treatment")
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tm_change = ((reformat_tumor_mean.iloc[-1] - reformat_tumor_mean.iloc[0])/reformat_tumor_mean.iloc[0]) * 100

# Display the data to confirm
tm_change
```

```python
# Store all Relevant Percent Changes into a Tuple
pct_change = (tm_change['Capomulin'], tm_change['Infubinol'], tm_change['Ketapril'], tm_change['Placebo'])
x_axis = np.arange(len(pct_change))

# Splice the data between passing and failing drugs
passed = plt.bar(x_axis[0], pct_change[0], width = -1, align = 'edge', color='green')
failed_1 = plt.bar(x_axis[1], pct_change[1], width = -1, align = 'edge', color='red')
failed_2 = plt.bar(x_axis[2], pct_change[2], width = -1, align = 'edge', color='red')
failed_3 = plt.bar(x_axis[3], pct_change[3], width = -1, align = 'edge', color='red')

# Orient widths. Add labels, tick marks, etc.
plt.title('Tumor Change Over 45 Day Treatment')
plt.ylabel('% Tumor Volume Change')
plt.xticks(x_axis,["Capomulin","Infubinol", "Ketapril","Placebo"])
plt.grid()

# Use functions to label the percentages of changes
def autolabelpass(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
    plt.annotate('{:,.0f}%'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, -8),
                    ha='center', va='bottom', color='white')

def autolabelfail(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
    plt.annotate('{:,.0f}%'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, 8),
                    ha='center', va='top', color='white')

# Call functions to implement the function calls
autolabelpass(passed)
autolabelfail(failed_1)
autolabelfail(failed_2)
autolabelfail(failed_3)

# Save the Figure
plt.savefig("Tumor Change Over 45 Day Treatment")

# Show the Figure
plt.show()
```
